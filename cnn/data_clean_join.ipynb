### 1. SOFAS - CATEGORY 1

```python
# category 1 - sofas
sofa_images= pd.read_csv('../data/scrape/image_data_sofa.csv')
sofa_images = sofa_images.drop(columns='Unnamed: 0')
sofa_links = pd.read_csv('../data/scrape/product_links_sofa.csv')
sofa_links = sofa_links.drop(columns='Unnamed: 0')
sofa_details = pd.read_csv('../data/scrape/product_details_sofa.csv')
sofa_details = sofa_details.drop(columns='Unnamed: 0')

sofa_images['label'] = 'sofa'
sofa_images['background'] = 0
sofa_images['delete'] = 0

sofa = pd.merge(sofa_images,sofa_details,\
                how='left',left_on='page_link',right_on='page_link')
sofa = sofa.drop(columns= 'product_name_y')
sofa.rename(columns={'product_name_x': 'product_name'}, inplace=True)
```

```python
print(sofa.shape)
print(sofa['image_link'].nunique())

sofa= sofa.drop_duplicates(subset='image_link')
print(sofa.shape)
```

```python
sofa.head(5)
```

```python
# Send to clean manually
sofa.to_csv('../data/scrape/clean_it_sofa.csv')
```

```python
# Read manually cleaned file
sofa_cleaned = pd.read_csv('../data/scrape/cleaned_sofa_complete.csv')
sofa_cleaned = sofa_cleaned.drop(columns ='Unnamed: 0')
print(sofa_cleaned.shape)
```
